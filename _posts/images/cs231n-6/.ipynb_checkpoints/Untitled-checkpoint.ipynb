```python
import os

def rename_file(): # 현재 위치(.)의 파일을 모두 가져온다.
    for filename in os.listdir("."): # 파일 확장자가 (properties)인 것만 처리
        if filename.endswith("jpg"): # 파일명에서 AA를 BB로 변경하고 파일명 수정
            new_filename = filename.replace("cs231n_2017_lecture6_페이지_", "")
            
            os.rename(filename, new_filename)
```

```python
rename_file()
```
